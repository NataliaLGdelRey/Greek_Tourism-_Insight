```python
# Import the modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

```

```python
# Read the CSV file from the Resources folder and create a Pandas DataFrame
Greece_data_df = pd.read_csv("Resources/travel_tourism_dataset.csv")
```

```python
# Review the DataFrame showing the first five rows
Greece_data_df.head()
```

```python
# Show how what mode of transportation is used the most to least
unique_modes_of_travel = Greece_data_df['Mode of Travel'].value_counts()
print(unique_modes_of_travel)
```

```python
# Show the dates of travel from least traveled date to most traveled date
dates = Greece_data_df['Date of Travel'].value_counts().sort_values(ascending=True)
print(dates)

```

```python
#show the number of people that traveled by date
grouped_date = Greece_data_df.groupby('Date of Travel')
date_counts = grouped_date.size()
print(date_counts)
```

```python
# Convert 'Date of Travel' column in datetime format
Greece_data_df['Date of Travel'] = pd.to_datetime(Greece_data_df['Date of Travel'])

# Create a new column 'Month' that extracts the month from 'Date of Travel'
Greece_data_df['Month'] = Greece_data_df['Date of Travel'].dt.strftime('%Y-%m')

# group the data by month
Greece_counts = Greece_data_df['Month'].value_counts().sort_index()
Greece_counts.head()
```

```python
#show the dataframe
Greece_data_df.head()
```

```python
#create new column to name the month they traveled
Greece_data_df['Name of the Month'] = Greece_data_df['Date of Travel'].dt.strftime('%Y-%m')
#change the name of the month to the actual month name
Greece_data_df['Name of the Month'] = pd.to_datetime(Greece_data_df['Name of the Month']).dt.strftime('%B')
#show the dataframe
Greece_data_df.head()
```

```python
#create a new column for the year they traveled
Greece_data_df['Year'] = Greece_data_df['Date of Travel'].dt.strftime('%Y')
#show the dataframe
Greece_data_df.head()
```

```python
#show the total number of people that traveled by Car in 2021
car_2021 = Greece_data_df.loc[(Greece_data_df['Mode of Travel'] == 'Car') & (Greece_data_df['Year'] == '2021')]
car_2021_count = car_2021['Mode of Travel'].count()
#seperate them by month
car_2021_month = car_2021.groupby('Name of the Month')
car_2021_month_count = car_2021_month['Mode of Travel'].count()
#show the dataframe
car_2021_month_count.head(12)
```

```python
#show the total number of people that traveled by Car in 2022
car_2022 = Greece_data_df.loc[(Greece_data_df['Mode of Travel'] == 'Car') & (Greece_data_df['Year'] == '2022')]
car_2022_count = car_2022['Mode of Travel'].count()
#seperate them by month
car_2022_month = car_2022.groupby('Name of the Month')
car_2022_month_count = car_2022_month['Mode of Travel'].count()
#show the dataframe
car_2022_month_count.head(12)
```

```python
# Export cleaned data to csv
Greece_data_df.to_csv("cleaned_data_AR.csv", index=False)
```
